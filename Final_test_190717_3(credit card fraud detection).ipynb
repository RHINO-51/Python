```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import mglearn
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
```

```python
card_df = pd.read_csv('creditcard.csv')
```

```python
card_df.head()
```

```python
card_df.info()
```

```python
card_df.hist(bins=50, figsize=(20,15))
```

```python
pca = PCA(n_components=2).fit(card_df)
df_pca = pca.transform(card_df)
df_pca.shape
```

KM

```python
km = KMeans(n_clusters=2).fit(df_pca)
group =  km.predict(df_pca)
```

```python
l = mglearn.discrete_scatter(df_pca[:,0],df_pca[:,1], group)
```

db

```python
p = DBSCAN(eps=50, min_samples=100).fit_predict(df_pca)  
l = mglearn.discrete_scatter(df_pca[:,0],df_pca[:,1],p)
```

agg

```python
agg_model = AgglomerativeClustering(n_clusters=2).fit(df_pca)
```

```python
p = agg_model.fit_predict(df_pca) #predict
l = mglearn.discrete_scatter(df_pca[:,0],df_pca[:,1], p)
```

```python
from sklearn.tree import DecisionTreeClassifier
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
import graphviz
from sklearn.tree import export_graphviz
```

```python

X = card_df.drop(columns='Class')
y = card_df['Class']
```

```python
model = DecisionTreeClassifier(max_depth=5).fit(X,y)
```

```python
model.score(X,y)
```

```python
export_graphviz(model, out_file='card_fraud.dot', class_names=['card_fraud','card_correct'],
                feature_names=X.columns, impurity=False, filled=True)
with open('card_fraud.dot') as f : 
    dot_graph = f.read()
g = graphviz.Source(dot_graph)
display(g)
```
