```python
df_hot = pd.get_dummies(df)
pca = PCA(n_components=2).fit(df_hot)
df_pca = pca.transform(df_hot)

df_hot.shape, df_pca.shape

#l = mglearn.discrete_scatter(x[:,0],x[:,1],p)
```

KM

```python
km = KMeans(n_clusters=2).fit(df_pca) #x를 3개로 그룹화(cluster)
group = km.predict(df_pca)
```

```python
group
```

```python
l = mglearn.discrete_scatter(df_pca[:,0],df_pca[:,1], group) #예측
```

db

```python
p = DBSCAN(eps=50, min_samples=50).fit_predict(df_pca) #k 개를 정해주지 않습니다. 
l = mglearn.discrete_scatter(df_pca[:,0],df_pca[:,1],p)
```

agg

```python
model = AgglomerativeClustering(n_clusters=2).fit(df_hot)
```

```python
p = model.fit_predict(df_hot) #predict
p, len(p)
```

```python
l = mglearn.discrete_scatter(df_pca[:,0],df_pca[:,1], p)
```

```python
from sklearn.tree import DecisionTreeClassifier
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
import graphviz
from sklearn.tree import export_graphviz
```

```python
df_hot.head()
df = df.replace('yes', 1)
df = df.replace('no', 0)
df_hot = pd.get_dummies(df)
df_hot['y'].mean()
```

```python
X = df_hot.drop(columns='y')
y = df_hot['y']
```

```python
model = DecisionTreeClassifier(max_depth=5).fit(X,y)
```

```python
model.score(X,y)
```

```python
export_graphviz(model, out_file='loan.dot', class_names=['Rent OK','Rent NO'],
                feature_names=X.columns, impurity=False, filled=True)
with open('loan.dot') as f : 
    dot_graph = f.read()
g = graphviz.Source(dot_graph)
display(g)
```
